# Figure Object-Part Two:Understanding The Figure

```python
a=np.linspace(0,10,11)
```

```python
b=a**4
```

```python
a
```

```python
b
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
fig=plt.figure()

#Large Axes

axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')



#Small Axes
axes2=fig.add_axes([0.2,0.5,0.25,0.25])  #[1,1,2,2]
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')


plt.show()
```

# Subplots

```python
a =np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)


for ax in axes:
    ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)

axes[0].plot(x,y)

plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

fig.subplots_adjust(wspace=1,hspace=0.5)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,2')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')


fig.suptitle("Figure Level",fontsize=16)

plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,2')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')


fig.suptitle("Figure Level",fontsize=16)

plt.tight_layout()

fig.savefig('new_subplots.png',bbox_inches='tight')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,2')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')


fig.suptitle("Figure Level",fontsize=16)

plt.tight_layout()

fig.set_figwidth(10)

fig.savefig('new_subplots.png',bbox_inches='tight')
```

# Styling-Part One:Legends

```python
x=np.linspace(0,11,10)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(1.1,0.5))
```

# Styling-Part Two:Colors and Styles

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#650dbd')  #rgb Hex Code 
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=10)  #rgb Hex Code 

ax.plot(x,x+1,color='purple',linewidth=5)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=10,ls='--')  #rgb Hex Code 

ax.plot(x,x+1,color='purple',linewidth=5,linestyle='-.')
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#e6c017',lw=10,ls='--')  #rgb Hex Code 

lines=ax.plot(x,x+1,color='purple',linewidth=5)
lines[0].set_dashes([1,1,1,2,3,5])
```
