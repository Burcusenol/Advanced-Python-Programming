# Combining DataFrames-Part One-Concatenation

```python
import numpy as np
import pandas as pd
```

```python
data_one={'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}
```

```python
data_two={'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}
```

```python
one=pd.DataFrame(data_one)
```

```python
two=pd.DataFrame(data_two)
```

```python
one
```

```python
two
```

```python
pd.concat([one,two],axis=1)
```

```python
pd.concat([two,one],axis=1)
```

```python
pd.concat([one,two],axis=0)
```

```python
two.columns=one.columns
```

```python
two
```

```python
mydf=pd.concat([one,two],axis=0)
```

# Combining DataFrames-Part Two-Inner Merge

```python
registirations=pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Clarie','David']})
logins=pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registirations
```

```python
logins
```

```python
pd.merge(registirations,logins,how='inner',on='name')
```

```python
pd.merge(logins,registirations,how='inner',on='name')
```

# Combining DataFrames-Part Three-Left and Right Merge

```python
pd.merge(left=registirations,right=logins,how='right',on='name')
```

```python
pd.merge(left=registirations,right=logins,how='left',on='name')
```

# Combining DataFrames-Part Four-Outer Merge

```python
pd.merge(registirations,logins,how='outer',on='name')
```

```python
registirations=registirations.set_index('name')
```

```python
registirations
```

```python
pd.merge(registirations,logins,left_index=True,right_on='name',how='inner')
```

```python
registirations=registirations.reset_index()
```

```python
registirations
```

```python
registirations.columns=['reg_name','reg_id']
```

```python
registirations
```

```python
logins
```

```python
results=pd.merge(registirations,logins,how='inner',left_on='reg_name',right_on='name')
```

```python
results.drop('reg_name',axis=1)
```

```python
registirations.columns=['name','id']
```

```python
logins.columns=['id','name']
```

```python
registirations
```

```python
logins
```

```python
pd.merge(registirations,logins,how='inner',on='name')
```

```python
pd.merge(registirations,logins,how='inner',on='name',suffixes=('_reg','_log'))
```

# Pivot Tables

```python
df=pd.read_csv('Sales_Funnel_CRM.csv')
```

```python
df
```

```python
licenses=df[['Company','Product','Licenses']]
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

```python
pd.pivot_table(df,index="Company",aggfunc='sum')
```

```python
df.groupby('Company').sum()
```

```python
pd.pivot_table(df,index="Company",aggfunc='sum',values=["Licenses","Sale Price"])
```

```python
df
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],aggfunc="sum")
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc="sum",fill_value=0)
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc=[np.sum,np.mean],fill_value=0)
```

```python
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price"],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price"],aggfunc=[np.sum],fill_value=0,margins=True)
```
